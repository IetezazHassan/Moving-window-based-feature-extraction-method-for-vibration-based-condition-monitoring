```python
broken_df = pd.concat([tb_gearbox_0,tb_gearbox_10,tb_gearbox_20,tb_gearbox_30,tb_gearbox_40,tb_gearbox_50,tb_gearbox_60,tb_gearbox_70,tb_gearbox_80,tb_gearbox_90],axis=0,ignore_index=True)
```

```python
healthy_df = pd.concat([h_gearbox_0,h_gearbox_10,h_gearbox_20,h_gearbox_30,h_gearbox_40,h_gearbox_50,h_gearbox_60,h_gearbox_70,h_gearbox_80,h_gearbox_90],axis=0,ignore_index=True)
```

```python
gear_data   = pd.concat([broken_df,healthy_df], axis =0)
```

```python
training_features = ['a1', 'a2', 'a3', 'a4', 'load']
label = ['failure']
```

```python
x = gear_data[training_features]
x.shape
```

```python
y = gear_data[label]
y.shape 
```

```python
x,y = shuffle(x,y)
```

```python
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

# Decision Tree

```python
DT_Model = tree.DecisionTreeClassifier()
DT_Model.fit(x_train, y_train)
DT_Model.score(x_test, y_test)
```

```python
y_pred = DT_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR_Model = LogisticRegression()
LR_Model.fit(x_train, y_train)
LR_Model.score(x_test, y_test)
```

```python
y_pred = LR_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

RF_Model = RandomForestClassifier()
RF_Model.fit(x_train, y_train)
RF_Model.score(x_test, y_test)
```

```python
y_pred = RF_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

NB_Model = GaussianNB()
NB_Model.fit(x_train, y_train)
NB_Model.score(x_test, y_test)
```
