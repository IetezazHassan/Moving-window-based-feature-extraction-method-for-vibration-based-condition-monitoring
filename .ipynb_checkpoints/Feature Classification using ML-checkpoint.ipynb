```python
import numpy as np
import pandas as pd

from matplotlib import pyplot
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import NearestNeighbors
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
```

```python
h0_df = pd.read_csv('Extracted Features/Healthy/h30hz0_features.csv')
h1_df = pd.read_csv('Extracted Features/Healthy/h30hz10_features.csv')
h2_df = pd.read_csv('Extracted Features/Healthy/h30hz20_features.csv')
h3_df = pd.read_csv('Extracted Features/Healthy/h30hz30_features.csv')
h4_df = pd.read_csv('Extracted Features/Healthy/h30hz40_features.csv')
h5_df = pd.read_csv('Extracted Features/Healthy/h30hz50_features.csv')
h6_df = pd.read_csv('Extracted Features/Healthy/h30hz60_features.csv')
h7_df = pd.read_csv('Extracted Features/Healthy/h30hz70_features.csv')
h8_df = pd.read_csv('Extracted Features/Healthy/h30hz80_features.csv')
h9_df = pd.read_csv('Extracted Features/Healthy/h30hz90_features.csv')
```

```python
dataframes = [h0_df, h1_df, h2_df, h3_df, h4_df, h5_df, h6_df, h7_df, h8_df, h9_df]
combined_healthy_df = pd.concat(dataframes, ignore_index=True)
```

```python
b0_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz0_features.csv')
b1_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz10_features.csv')
b2_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz20_features.csv')
b3_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz30_features.csv')
b4_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz40_features.csv')
b5_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz50_features.csv')
b6_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz60_features.csv')
b7_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz70_features.csv')
b8_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz80_features.csv')
b9_df = pd.read_csv('Extracted Features/Broken Tooth/b30hz90_features.csv')
```

```python
dataframes = [b0_df, b1_df, b2_df, b3_df, b4_df, b5_df, b6_df, b7_df, b8_df, b9_df]
combined_faulty_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_faulty_df.shape
```

```python
b0_df.shape, b1_df.shape, b2_df.shape, b3_df.shape, b4_df.shape, b5_df.shape, b6_df.shape, b7_df.shape, b8_df.shape, b9_df.shape,
```

```python
full_dataset = [combined_healthy_df, combined_faulty_df]
final_df = pd.concat(full_dataset, ignore_index=True)
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
inputs = final_df.drop(columns=['label'])
outputs = final_df['label']
```

```python
inputs.shape, outputs.shape
```

```python
inputs, outputs = shuffle(inputs,outputs)
```

```python
inputs.shape, outputs.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, outputs, test_size=0.20, random_state=42)
```

```python
x_train.shape, x_test.shape
```

# Decision Tree

```python
DT_Model = tree.DecisionTreeClassifier()
DT_Model.fit(x_train, y_train)
DT_Model.score(x_test, y_test)
```

```python
y_pred = DT_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

RF_Model = RandomForestClassifier()
RF_Model.fit(x_train, y_train)
RF_Model.score(x_test, y_test)
```

```python
y_pred = RF_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

NB_Model = GaussianNB()
NB_Model.fit(x_train, y_train)
NB_Model.score(x_test, y_test)
```

```python
y_pred = NB_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR_Model = LogisticRegression()
LR_Model.fit(x_train, y_train)
LR_Model.score(x_test, y_test)
```

```python
y_pred = LR_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```

# SVM

```python
from sklearn import svm

SVM_Model = svm.SVC()
SVM_Model.fit(x_train, y_train)
SVM_Model.score(x_test, y_test)
```

```python
y_pred = SVM_Model.predict(x_test)
print(classification_report(y_pred, y_test))
```
