```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

from matplotlib import pyplot
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
def extract_time_domain_features(data):    
    # Mean
    mean = np.mean(data)
    
    # Root Mean Square (RMS)
    rms = np.sqrt(np.mean(np.square(data)))
    
    # Standard Deviation
    std_dev = np.std(data)
    
    # Variance
    variance = np.var(data)
    
    # Skewness
    skewness = skew(data)
    
    # Kurtosis
    kurtosis_f = kurtosis(data)
    
    # Peak-to-Peak (Peak Range)
    peak_to_peak = np.ptp(data)
    
    # Crest Factor (Peak / RMS)
    crest_factor = np.max(np.abs(data)) / rms
    
    # Shape Factor (RMS / Mean Absolute Value)
    shape_factor = rms / np.mean(np.abs(data))
    
    # Impulse Factor (Peak / Mean Absolute Value)
    impulse_factor = np.max(np.abs(data)) / np.mean(np.abs(data))
    
    return mean, rms, std_dev, variance, skewness, kurtosis_f, peak_to_peak, crest_factor, shape_factor, impulse_factor
```

```python
h_gearbox_0 = pd.read_csv('Gearbox Dataset/Healthy/h30hz0.csv')
```

```python
chunk_size = 50
num_rows = len(h_gearbox_0)
extracted_features = []


val = 0
for start in range(0, num_rows, chunk_size):
    chunk = h_gearbox_0[start:start + chunk_size]
    
    sensor_1 = chunk['a1']
    sensor_2 = chunk['a2']
    sensor_3 = chunk['a3']
    sensor_4 = chunk['a4']
    
    sensor_1_time_domain_features = extract_time_domain_features(sensor_1)
    sensor_2_time_domain_features = extract_time_domain_features(sensor_2)
    sensor_3_time_domain_features = extract_time_domain_features(sensor_3)
    sensor_4_time_domain_features = extract_time_domain_features(sensor_4)
    
    combined_features = sum((sensor_1_time_domain_features, sensor_2_time_domain_features, sensor_3_time_domain_features, sensor_4_time_domain_features), ())
    extracted_features.append(combined_features)

    
    #val = val + 1
    print(start, start+chunk_size) 
```
